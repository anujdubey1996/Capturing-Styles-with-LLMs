```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

```python
def get_embeddings(text):
    # Define your cURL command
    curl_command = """
    curl https://api.openai.com/v1/embeddings \
      -H "Authorization: Bearer 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" \
      -H "Content-Type: application/json" \
      -d '{
        "input": "{text}",
        "model": "text-embedding-ada-002",
        "encoding_format": "float"
          }'
    """

    # Run the command and capture the output
    process = subprocess.Popen(curl_command, stdout=subprocess.PIPE, stderr=subprocess.PIPE, shell=True)
    output, error = process.communicate()

    # Check for errors
    if process.returncode != 0:
        print("Error executing the cURL command:", error.decode())
    else:
        # Parse the JSON output
        json_data = json.loads(output.decode())
        return json_data
```

```python
# List of sentences in lowercase
sentences_lowercase = [
    "the quick brown fox jumps over the lazy dog",
    "all good things come to those who wait",
    "practice makes perfect",
    "knowledge is power",
    "to be or not to be, that is the question"
]

# Converting each sentence to uppercase
sentences_uppercase = [sentence.upper() for sentence in sentences_lowercase]

print("Lowercase Sentences:", sentences_lowercase)
print("Uppercase Sentences:", sentences_uppercase)
```

```python
lowercase_embeddings = []
for text in sentences_lowercase:
    json_data = get_embeddings(text)
    lowercase_embeddings.append(json_data['data'][0]['embedding'])


uppercase_embeddings = []
for text in sentences_uppercase:
    json_data = get_embeddings(text)
    uppercase_embeddings.append(json_data['data'][0]['embedding'])
```

```python

# Assuming 'embeddings_list_1' and 'embeddings_list_2' are your lists of embeddings
embeddings_array_1 = np.array(uppercase_embeddings)
embeddings_array_2 = np.array(lowercase_embeddings)

# Assuming 'labels_1' and 'labels_2' are the lists of labels corresponding to the embeddings
labels_2 = ['l1', 'l2', 'l3', 'l4', 'l5']
labels_1 = ['u1', 'u2', 'u3', 'u4', 'u5']

# Combine both arrays for t-SNE
combined_embeddings = np.concatenate((embeddings_array_1, embeddings_array_2))
combined_labels = labels_1 + labels_2  # Combine labels

# Use t-SNE to project the combined embeddings into 2D space
tsne = TSNE(n_components=2, random_state=0)
combined_embeddings_2d = tsne.fit_transform(combined_embeddings)

# Split the transformed embeddings and labels back into two lists
split_point = embeddings_array_1.shape[0]
embeddings_2d_1 = combined_embeddings_2d[:split_point]
embeddings_2d_2 = combined_embeddings_2d[split_point:]

# Plotting with labels and lines
plt.figure(figsize=(12, 8))
for i in range(len(embeddings_2d_1)):
    # Plot points
    plt.scatter(embeddings_2d_1[i, 0], embeddings_2d_1[i, 1], color='blue')
    plt.scatter(embeddings_2d_2[i, 0], embeddings_2d_2[i, 1], color='red')

    # Plot lines
    plt.plot([embeddings_2d_1[i, 0], embeddings_2d_2[i, 0]], [embeddings_2d_1[i, 1], embeddings_2d_2[i, 1]], color='grey', linestyle='--')

    # Plot labels
    plt.text(embeddings_2d_1[i, 0], embeddings_2d_1[i, 1], labels_1[i], fontsize=9)
    plt.text(embeddings_2d_2[i, 0], embeddings_2d_2[i, 1], labels_2[i], fontsize=9)

plt.title('2D projection of embeddings with labels and lines')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()
```
